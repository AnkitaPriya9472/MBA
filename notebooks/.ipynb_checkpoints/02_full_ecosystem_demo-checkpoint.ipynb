## 1. Setup

```python
import sys
sys.path.insert(0, '..')

import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Core simulation API
from simulation.engine import run_scenario
from output.results import SimulationResult

# Models (for type hints / inspection)
from models.agent_types import AgentProfile, AgentState, RoundResponse
from models.world_state import WorldState
from models.market import MarketState, TradeRoute
from models.scenario import ShockScenario
from models.events import WorldEvent

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 11

# Verify API key is set
assert os.getenv('OPENROUTER_API_KEY'), (
    'OPENROUTER_API_KEY is not set.\n'
    'Run: export OPENROUTER_API_KEY="your-key-here"'
)
print('✓ API key found')
```

## 2. Configure Paths

```python
SCENARIO_PATH      = '../scenarios/tariff_shock.json'
NETWORK_PATH       = '../network/steel_ecosystem.json'
RESULTS_PATH       = '../data/runs/full_ecosystem_results.json'

AGENT_PROFILE_PATHS = [
    '../agents/profiles/companies/tata_steel.json',
    '../agents/profiles/governments/india.json',
    '../agents/profiles/consumers/industrial_steel.json',
    '../agents/profiles/suppliers/iron_ore.json',
    '../agents/profiles/regulators/dgtr.json',
    '../agents/profiles/investors/fii_aggregate.json',
]

ROUNDS  = 3
MODEL   = 'moonshotai/kimi-k2'

# Quick preview: load scenario metadata without running
scenario = ShockScenario.from_json(SCENARIO_PATH)
print(f'Scenario  : {scenario.name}')
print(f'Category  : {scenario.category}')
print(f'Severity  : {scenario.severity}')
print(f'Sectors   : {', '.join(scenario.affected_sectors)}')
print(f'Context   : {scenario.context[:200]}...')
```

## 3. Run the Simulation

`run_scenario()` is the single convenience entrypoint. Internally it:
1. Loads all JSON files (scenario, agent profiles, trade network)
2. Initialises 5 sub-engines: `MarketClearing`, `TradeNetwork`, `CaptureEngine`, `SentimentEngine`, `StateResolver`
3. Runs the **4-phase round loop** for each round:
   - Phase 1 — Build `RoundContext` per agent (visibility filtering + capture bias injection)
   - Phase 2 — Execute all agents in parallel (async LLM calls)
   - Phase 3 — `StateResolver.resolve()` → market clearing, trade recalculation, sentiment update
   - Phase 4 — Apply patches to `WorldState`, fire events
4. Returns a `SimulationResult` with a full `WorldState` snapshot

```python
result: SimulationResult = run_scenario(
    scenario_path=SCENARIO_PATH,
    agent_profile_paths=AGENT_PROFILE_PATHS,
    network_path=NETWORK_PATH,
    rounds=ROUNDS,
    model=MODEL,
)

print(f'\nRounds completed : {result.rounds_completed}')
print(f'Model used       : {result.metadata["model_used"]}')
print(f'Duration         : {result.metadata["duration_seconds"]:.1f}s')
print(f'Total API calls  : {result.metadata["total_api_calls"]}')
print(f'Fallbacks        : {result.metadata["total_fallbacks"]}')
print(f'Events fired     : {len(result.get_events())}')
```

## 4. Agent Decision Timeline

Each agent produces a `RoundResponse` per round containing:
- `primary_action` — constrained to role-specific action enum
- `secondary_action` — optional supporting move
- `assessment` — LLM narrative reasoning (max 500 chars)
- `confidence` — self-reported confidence (0–1)
- `metrics_change` — quantitative predictions
- `is_fallback` — True if the LLM failed and a safe default was used

```python
rows = []
for round_num in range(1, result.rounds_completed + 1):
    for agent_id, resp in result.get_round(round_num).items():
        profile = result.world_state.agent_profiles[agent_id]
        rows.append({
            'Round'          : round_num,
            'Agent'          : profile.name,
            'Type'           : profile.agent_type.value,
            'Primary Action' : resp.primary_action,
            'Secondary'      : resp.secondary_action or '—',
            'Confidence'     : f'{resp.confidence:.0%}',
            'Fallback?'      : '⚠' if resp.is_fallback else 'No',
            'Assessment'     : resp.assessment[:120] + '…' if len(resp.assessment) > 120 else resp.assessment,
        })

df_timeline = pd.DataFrame(rows)
pd.set_option('display.max_colwidth', 130)
df_timeline
```

## 5. Confidence Trajectory

Confidence typically **decreases** as uncertainty compounds across rounds and agents react to each other's moves.

```python
conf_rows = []
for round_num in range(1, result.rounds_completed + 1):
    for agent_id, resp in result.get_round(round_num).items():
        name = result.world_state.agent_profiles[agent_id].name
        conf_rows.append({'Round': round_num, 'Agent': name, 'Confidence': resp.confidence})

df_conf = pd.DataFrame(conf_rows)
pivot = df_conf.pivot(index='Round', columns='Agent', values='Confidence')

fig, ax = plt.subplots()
for col in pivot.columns:
    ax.plot(pivot.index, pivot[col], marker='o', label=col)
ax.set_title('Agent Confidence per Round')
ax.set_xlabel('Round')
ax.set_ylabel('Confidence')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))
ax.set_xticks(range(1, result.rounds_completed + 1))
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=9)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(pivot.map(lambda x: f'{x:.0%}').to_string())
```

## 6. Market Clearing — Price & Supply/Demand

After every round the `MarketClearing` engine runs a double-auction:
- Aggregates supply bids and demand offers from agent actions
- Applies tariff-adjusted price discovery
- Tracks `unmet_demand_pct` as a stress indicator

```python
# Current market states after simulation
market_rows = []
for commodity, mkt in result.world_state.market_states.items():
    market_rows.append({
        'Commodity'       : commodity,
        'Spot Price'      : f'${mkt.spot_price:,.0f}',
        'Prev Price'      : f'${mkt.prev_price:,.0f}' if mkt.prev_price else '—',
        'Price Δ%'        : f'{mkt.price_change_pct:+.1f}%' if mkt.price_change_pct else '—',
        'Demand (MT)'     : f'{mkt.demand:,.0f}',
        'Supply (MT)'     : f'{mkt.supply:,.0f}',
        'Cleared (MT)'    : f'{mkt.volume_cleared:,.0f}',
        'Unmet Demand %'  : f'{mkt.unmet_demand_pct:.1f}%',
    })

df_market = pd.DataFrame(market_rows)
print('Final Market State (after last round):')
df_market
```

```python
# Price history chart
if not result.world_state.market_states:
    print('No market state data available — market clearing may not have run.')
else:
    fig, axes = plt.subplots(1, len(result.world_state.market_states), squeeze=False)
    axes = axes[0]

    for ax, (commodity, mkt) in zip(axes, result.world_state.market_states.items()):
        if mkt.price_history:
            ax.plot(range(len(mkt.price_history)), mkt.price_history, marker='o', color='steelblue')
            ax.set_title(f'{commodity} price history')
            ax.set_xlabel('Tick')
            ax.set_ylabel('USD/MT')
            ax.grid(alpha=0.3)
        else:
            ax.set_title(f'{commodity}')
            ax.text(0.5, 0.5, 'No price history', ha='center', va='center', transform=ax.transAxes)

    plt.suptitle('Commodity Price Evolution', fontweight='bold')
    plt.tight_layout()
    plt.show()
```

## 7. Trade Route Analysis

The `TradeNetwork` tracks bilateral trade flows. Each route has:
- `tariff_rate` — updated live when governments impose/remove tariffs
- `friction` — non-tariff barriers (capital controls, logistics)
- `volume` — MT flowing this round

```python
trade_rows = []
for route in result.world_state.trade_routes:
    trade_rows.append({
        'Route'         : f'{route.from_country} → {route.to_country}',
        'Commodity'     : route.commodity,
        'Volume (MT)'   : f'{route.volume:,.0f}',
        'Tariff Rate %' : f'{route.tariff_rate:.1f}%',
        'Friction'      : f'{route.friction:.2f}',
    })

df_trade = pd.DataFrame(trade_rows)
print('Active Trade Routes (final state):')
df_trade
```

```python
# Bar chart of trade volumes by route
if trade_rows:
    df_trade_plot = pd.DataFrame(trade_rows)
    df_trade_plot['Volume_num'] = [r.volume for r in result.world_state.trade_routes]
    labels = df_trade_plot['Route'] + '\n(' + df_trade_plot['Commodity'] + ')'

    fig, ax = plt.subplots()
    bars = ax.bar(labels, df_trade_plot['Volume_num'], color='steelblue', alpha=0.8)
    ax.bar_label(bars, fmt='{:,.0f} MT', padding=3, fontsize=9)
    ax.set_title('Trade Volume by Route (final round)')
    ax.set_ylabel('Volume (MT)')
    ax.tick_params(axis='x', labelsize=9)
    ax.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## 8. Agent Sentiment — Fear-Greed Index

The `SentimentEngine` updates each agent's `fear_greed_index` (0 = extreme fear, 10 = extreme greed) each round using:
- **Prospect theory** — losses weigh ~2× more than equal gains (Kahneman-Tversky)
- **Herding** — agents partially follow peer sentiment
- **Momentum** + regression to mean

```python
sentiment_rows = []
for agent_id, state in result.world_state.agent_states.items():
    profile = result.world_state.agent_profiles[agent_id]
    sentiment_rows.append({
        'Agent'            : profile.name,
        'Type'             : profile.agent_type.value,
        'Fear-Greed (0-10)': round(state.fear_greed_index, 2),
        'Sentiment'        : '😨 Fear' if state.fear_greed_index < 4 else
                             '😐 Neutral' if state.fear_greed_index < 6 else '😄 Greed',
        'Financial Health' : f'{state.financial_health:.0%}',
        'Capture Score'    : f'{state.capture_score:.2f}',
    })

df_sentiment = pd.DataFrame(sentiment_rows).sort_values('Fear-Greed (0-10)')
df_sentiment
```

```python
# Horizontal bar chart
fig, ax = plt.subplots()
colors = ['#d62728' if v < 4 else '#aec7e8' if v < 6 else '#2ca02c'
          for v in df_sentiment['Fear-Greed (0-10)']]
bars = ax.barh(df_sentiment['Agent'], df_sentiment['Fear-Greed (0-10)'], color=colors, alpha=0.85)
ax.axvline(x=5, color='gray', linestyle='--', alpha=0.6, label='Neutral (5.0)')
ax.bar_label(bars, fmt='%.2f', padding=3, fontsize=10)
ax.set_xlim(0, 10)
ax.set_xlabel('Fear-Greed Index (0=Extreme Fear, 10=Extreme Greed)')
ax.set_title('Agent Sentiment After Final Round')
ax.legend()
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 9. World Events

The `StateResolver` fires `WorldEvent` objects when threshold conditions are met. Key event types:

| Event Type | Trigger |
|---|---|
| `MARKET_FLOOD` | A trade route volume surged >40% in one round |
| `TRADE_WAR_ESCALATION` | Both sides imposed retaliatory tariffs |
| `CAPTURE_TIPPING_POINT` | Regulator's capture score crossed 0.7 |
| `PRICE_SPIKE` | Commodity price rose >15% in one round |

```python
events = result.get_events()

if not events:
    print('No threshold events fired during this simulation.')
    print('(This means no trade route surged >40%, no capture tipping point was reached, etc.)')
else:
    event_rows = []
    for e in events:
        event_rows.append({
            'Round'           : e.round,
            'Event Type'      : e.event_type,
            'Description'     : e.description,
            'Affected Agents' : ', '.join(e.affected_agents),
        })
    df_events = pd.DataFrame(event_rows)
    print(f'Total events fired: {len(events)}')
    df_events
```

## 10. Regulatory Capture — DGTR Score

The `CaptureEngine` tracks how much lobbying pressure has accumulated on each regulator.
- Lobby actions from company/government agents increment capture score
- Score decays each round without reinforcement
- At score ≥ 0.7 → `CAPTURE_TIPPING_POINT` event fires and bias is injected into the regulator's next prompt

```python
# Show regulator agents and their capture state
from models.agent_types import AgentType

regulator_rows = []
for agent_id, profile in result.world_state.agent_profiles.items():
    if profile.agent_type == AgentType.regulator:
        state = result.world_state.agent_states[agent_id]
        regulator_rows.append({
            'Regulator'      : profile.name,
            'Capture Score'  : round(state.capture_score, 3),
            'Captured?'      : '⚠ YES' if state.capture_score >= 0.7 else '✓ Independent',
            'Financial Health': f'{state.financial_health:.0%}',
        })

if regulator_rows:
    df_capture = pd.DataFrame(regulator_rows)
    df_capture
else:
    print('No regulator agents in this simulation.')
```

```python
# Show DGTR's action history across rounds
dgtr_id = None
for agent_id, profile in result.world_state.agent_profiles.items():
    if profile.agent_type == AgentType.regulator:
        dgtr_id = agent_id
        break

if dgtr_id:
    history = result.get_agent_history(dgtr_id)
    print(f'DGTR ({dgtr_id}) decision history:')
    print()
    for resp in history:
        print(f'  Round {resp.round}: [{resp.primary_action}] (confidence: {resp.confidence:.0%})')
        print(f'  Assessment: {resp.assessment[:200]}')
        print()
```

## 11. Public Signal Network

Agents can broadcast public signals to influence others. These are **visible** to other agents (subject to `VisibilityFilter`).
- `to_agent=None` → broadcast to all
- `to_agent='some_id'` → directed signal

Information asymmetry: cross-border signals are delayed 1 round; domestic signals are instant.

```python
all_signals = result.world_state.all_signals

if not all_signals:
    print('No public signals were broadcast in this simulation.')
else:
    signal_rows = []
    for sig in all_signals:
        from_name = result.world_state.agent_profiles.get(sig.from_agent, {}).name if sig.from_agent in result.world_state.agent_profiles else sig.from_agent
        to_name = result.world_state.agent_profiles.get(sig.to_agent, {}).name if sig.to_agent and sig.to_agent in result.world_state.agent_profiles else ('ALL' if not sig.to_agent else sig.to_agent)
        signal_rows.append({
            'Round'  : sig.round,
            'From'   : from_name,
            'To'     : to_name,
            'Type'   : sig.signal_type,
            'Content': sig.content[:100] + '…' if len(sig.content) > 100 else sig.content,
        })

    df_signals = pd.DataFrame(signal_rows)
    print(f'Total public signals broadcast: {len(all_signals)}')
    df_signals
```

## 12. Private Intent — Observer-Only View

`PrivateIntent` is **never shared** with other agents. Only the simulation observer (this notebook) can read it.
This reveals hidden agendas — the gap between public signals and private reasoning.

```python
print('Private Intents (never visible to other agents):')
print('=' * 70)

any_found = False
for round_num in range(1, result.rounds_completed + 1):
    for agent_id, resp in result.get_round(round_num).items():
        if resp.private_intent:
            any_found = True
            name = result.world_state.agent_profiles[agent_id].name
            target = resp.private_intent.target_agent or 'undisclosed'
            print(f'[Round {round_num}] {name} (target: {target})')
            print(f'  → {resp.private_intent.content}')
            print()

if not any_found:
    print('No private intents recorded in this run.')
```

## 13. Agent History Deep-Dive

Use `result.get_agent_history(agent_id)` to trace any single agent's full strategy evolution.

```python
# Pick the first company agent
focus_agent_id = next(
    (aid for aid, p in result.world_state.agent_profiles.items()
     if p.agent_type.value == 'company'),
    list(result.world_state.agent_profiles.keys())[0]
)
focus_profile = result.world_state.agent_profiles[focus_agent_id]

print(f'Deep-dive: {focus_profile.name} ({focus_profile.agent_type.value})')
print('=' * 70)

for resp in result.get_agent_history(focus_agent_id):
    print(f'\nRound {resp.round}')
    print(f'  Primary Action  : {resp.primary_action}')
    print(f'  Secondary Action: {resp.secondary_action or "—"}')
    print(f'  Confidence      : {resp.confidence:.0%}')
    print(f'  Assessment      : {resp.assessment}')
    if resp.metrics_change:
        print(f'  Predicted Metrics:')
        for k, v in resp.metrics_change.items():
            print(f'    {k}: {v}')
    if resp.is_fallback:
        print('  ⚠ FALLBACK RESPONSE')
```

## 14. Quantitative Outcomes Summary

Aggregate the `metrics_change` predictions from the final round across all agents.

```python
print(f'Final Round ({result.rounds_completed}) — Predicted Metrics by Agent')
print('=' * 70)

final_round_responses = result.get_round(result.rounds_completed)
for agent_id, resp in final_round_responses.items():
    name = result.world_state.agent_profiles[agent_id].name
    print(f'\n{name}  [{resp.primary_action}]')
    if resp.metrics_change:
        for metric, val in resp.metrics_change.items():
            print(f'  {metric}: {val}')
    else:
        print('  (no quantitative metrics provided)')
```

## 15. Save & Reload Results

```python
# Save to disk
result.save(RESULTS_PATH)
print(f'Saved to: {RESULTS_PATH}')

# Reload and verify round count matches
loaded = SimulationResult.load(RESULTS_PATH)
assert loaded.rounds_completed == result.rounds_completed
assert loaded.scenario.name == result.scenario.name
print(f'Reloaded successfully — rounds: {loaded.rounds_completed}, scenario: {loaded.scenario.name}')
```
