**Domain:** Indian Manufacturing Sector  
**Scenario:** US imposes additional 25% tariff on steel imports from India  

---

## Overview

This notebook demonstrates a **multi-agent AI simulation** where distinct AI agents вЂ” representing real business ecosystem actors вЂ” respond dynamically to an external shock over multiple rounds of strategic interaction.

### Agents in this simulation:
1. **Tata Steel** вЂ” India's largest steel producer (company agent)
2. **Indian Government** вЂ” Ministry of Steel & Commerce (government agent)
3. **Steel Consumers** вЂ” Auto & Construction sectors (consumer agent)
4. **Raw Material Suppliers** вЂ” Iron ore & coking coal (supplier agent)

### Simulation flow:
- **Round 1:** Each agent reacts to the shock independently
- **Round 2:** Agents see Round 1 decisions and adjust strategies
- **Round 3:** Final strategic positions after two rounds of ecosystem interaction

## Setup

```python
import sys
sys.path.insert(0, '..')

from simulation.engine import run_scenario, SimulationEngine, ShockScenario
from agents.base import BaseAgent, AgentProfile
```

## Load Scenario & Agent Profiles

```python
# Define paths
SCENARIO_PATH = '../scenarios/tariff_shock.json'
PROFILE_PATHS = [
    '../agents/profiles/tata_steel.json',
    '../agents/profiles/indian_government.json',
    '../agents/profiles/steel_consumer.json',
    '../agents/profiles/iron_ore_supplier.json',
]

# Load scenario
scenario = ShockScenario.from_json(SCENARIO_PATH)
print(f'Scenario: {scenario.name}')
print(f'Severity: {scenario.severity}')
print(f'Category: {scenario.category}')
print(f'Affected sectors: {", ".join(scenario.affected_sectors)}')
```

```python
# Load agents
agents = [
    BaseAgent(profile=AgentProfile.from_json(p), model='moonshotai/kimi-k2')
    for p in PROFILE_PATHS
]

for agent in agents:
    print(f'  Loaded: {agent.profile.name} ({agent.profile.role})')
```

## Run Simulation (3 Rounds)

Each round, every agent:
1. Reads the shock event
2. Reviews other agents' previous decisions
3. Makes a strategic decision grounded in real company data
4. Explains its reasoning

```python
engine = SimulationEngine(scenario=scenario, agents=agents, rounds=3)
result = engine.run()
```

## Analysis: Agent Decision Timeline

Let's examine how each agent's strategy evolved across rounds.

```python
import pandas as pd

# Build a DataFrame of all responses
rows = []
for r in result.responses:
    rows.append({
        'Agent': r.agent_name,
        'Round': r.round_number,
        'Decision': r.decision,
        'Reasoning': r.reasoning,
        'Confidence': f'{r.confidence:.0%}',
        'Impact Areas': ', '.join(r.impact_areas),
    })

df = pd.DataFrame(rows)
df
```

## Confidence Trajectory

How does each agent's confidence in their strategy change as they see others' moves?

```python
# Confidence over rounds
confidence_data = []
for r in result.responses:
    confidence_data.append({
        'Agent': r.agent_name,
        'Round': r.round_number,
        'Confidence': r.confidence,
    })

df_conf = pd.DataFrame(confidence_data)
pivot = df_conf.pivot(index='Round', columns='Agent', values='Confidence')
print('\nConfidence scores per round:')
print(pivot.to_string())
```

## Key Metrics Changes

What quantitative impacts did each agent predict?

```python
# Show final round metrics for each agent
final_round = result.get_round(3)

for r in final_round:
    print(f'\n{r.agent_name} вЂ” Final Position:')
    print(f'  Decision: {r.decision}')
    if r.metrics_change:
        print(f'  Predicted Metrics:')
        for metric, change in r.metrics_change.items():
            print(f'    {metric}: {change}')
```
