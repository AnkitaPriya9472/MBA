## Overview

The Strait of Hormuz is the world's most critical oil chokepoint — ~20 million barrels/day (20% of global supply) transit through it. India imports 88% of its crude oil, with ~40% sourced from Gulf nations via this strait. A military closure would trigger a **multi-sector economic crisis** across energy, agriculture, aviation, financial markets, and fiscal policy.

### Agents in this simulation:
1. **Indian Oil Corporation (IOC)** — India's largest refiner and fuel retailer (company agent)
2. **Indian Government (Ministry of Petroleum & Finance)** — Energy security and fiscal policy (government agent)
3. **Indian Aviation Sector (IndiGo & Air India)** — Fuel-cost-sensitive transport sector (company agent)
4. **Indian Agriculture & Fertilizer Sector** — Food security and input cost vulnerability (consumer agent)
5. **Reserve Bank of India (RBI)** — Monetary policy, inflation, and forex stability (regulator agent)

### Simulation flow:
- **Round 1:** Each agent reacts to the Hormuz closure shock independently
- **Round 2:** Agents see Round 1 decisions and adjust strategies based on ecosystem moves
- **Round 3:** Final strategic positions after two rounds of inter-agent adaptation

## Load Scenario & Agent Profiles

```python
SCENARIO_PATH = '../scenarios/hormuz_crisis.json'
NETWORK_PATH  = '../network/hormuz_energy_network.json'
RESULTS_PATH  = '../data/hormuz_crisis_results.json'

PROFILE_PATHS = [
    '../agents/profiles/indian_oil_corp.json',
    '../agents/profiles/indian_govt_energy.json',
    '../agents/profiles/indian_airlines.json',
    '../agents/profiles/indian_farmer_fertilizer.json',
    '../agents/profiles/rbi_central_bank.json',
]

ROUNDS = 3
MODEL  = 'moonshotai/kimi-k2'

# Quick preview — load scenario metadata without running the simulation
scenario = ShockScenario.from_json(SCENARIO_PATH)
print(f'Scenario: {scenario.name}')
print(f'Severity: {scenario.severity}')
print(f'Category: {scenario.category}')
print(f'Affected sectors: {", ".join(scenario.affected_sectors)}')
```

## Run Simulation (3 Rounds)

Each round, every agent:
1. Reads the Hormuz closure shock event
2. Reviews other agents' previous decisions
3. Makes a strategic decision grounded in real institutional data
4. Explains its reasoning with confidence scores and predicted metrics

```python
result: SimulationResult = run_scenario(
    scenario_path=SCENARIO_PATH,
    agent_profile_paths=PROFILE_PATHS,
    network_path=NETWORK_PATH,
    rounds=ROUNDS,
    model=MODEL,
)

print(f'\nRounds completed : {result.rounds_completed}')
print(f'Model used       : {result.metadata["model_used"]}')
print(f'Duration         : {result.metadata["duration_seconds"]:.1f}s')
print(f'Total API calls  : {result.metadata["total_api_calls"]}')
print(f'Fallbacks        : {result.metadata["total_fallbacks"]}')
```

```python
import pandas as pd

# Build a DataFrame of all responses
rows = []
for round_num in range(1, result.rounds_completed + 1):
    for agent_id, resp in result.get_round(round_num).items():
        name = result.world_state.agent_profiles[agent_id].name
        rows.append({
            'Agent': name,
            'Round': resp.round,
            'Decision': resp.primary_action,
            'Reasoning': resp.assessment,
            'Confidence': f'{resp.confidence:.0%}',
            'Impact Areas': ', '.join(resp.impact_areas),
        })

df = pd.DataFrame(rows)
df
```

## Confidence Trajectory

How does each agent's confidence in their strategy change as they observe ecosystem-wide responses?

```python
# Confidence over rounds
confidence_data = []
for round_num in range(1, result.rounds_completed + 1):
    for agent_id, resp in result.get_round(round_num).items():
        name = result.world_state.agent_profiles[agent_id].name
        confidence_data.append({
            'Agent': name,
            'Round': resp.round,
            'Confidence': resp.confidence,
        })

df_conf = pd.DataFrame(confidence_data)
pivot = df_conf.pivot(index='Round', columns='Agent', values='Confidence')
print('\nConfidence scores per round:')
print(pivot.to_string())
```

## Key Metrics Changes

What quantitative impacts did each agent predict in their final-round positions?

```python
# Show final round metrics for each agent
final_round = result.get_round(result.rounds_completed)

for agent_id, resp in final_round.items():
    name = result.world_state.agent_profiles[agent_id].name
    print(f'\n{name} — Final Position:')
    print(f'  Decision: {resp.primary_action}')
    if resp.metrics_change:
        print(f'  Predicted Metrics:')
        for metric, change in resp.metrics_change.items():
            print(f'    {metric}: {change}')
```

## Cross-Agent Impact Matrix

How do decisions by one agent cascade to others? This is especially critical in the Hormuz scenario where energy, monetary policy, agriculture, and aviation are deeply interconnected.

```python
# Show each agent's decision per round in a compact view
for round_num in range(1, result.rounds_completed + 1):
    print(f'\n{"=" * 80}')
    print(f'ROUND {round_num}')
    print(f'{"=" * 80}')
    for agent_id, resp in result.get_round(round_num).items():
        name = result.world_state.agent_profiles[agent_id].name
        print(f'\n  [{name}] (Confidence: {resp.confidence:.0%})')
        # Truncate decision for readability
        decision_short = resp.primary_action[:200] + '...' if len(resp.primary_action) > 200 else resp.primary_action
        print(f'  Decision: {decision_short}')
```
